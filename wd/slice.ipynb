```python
m = 10
g = -10
key = f'm{m:04.1f}_g{g:+04.2f}_0'
s1 = pd.read_csv(f'csv/{key}/m{m}_final_slice.csv')
s1
```

```python
m = 11
g = -10
key = f'm{m:04.1f}_g{g:+04.2f}_0'
s2 = pd.read_csv(f'csv/{key}/m{m}_final_slice.csv')
s2
```

```python
r1 = np.array(s1['r (Rsun)'])
m1 = np.array(s1['mass enclosed (Msun)'])
r2 = np.array(s2['r (Rsun)'])
m2 = np.array(s2['mass enclosed (Msun)'])
plt.plot(r1, m1)
plt.plot(r2, m2)
plt.ylabel('Mass enclosed (Msun)')
plt.xlabel('Radius (Rsun)')
plt.title(f'Mass profile')
plt.scatter(r1[-1], m1[-1], label=f'10 Msun progenitor: ({r1[-1]:.6f} Rsun, {m1[-1]:.6f} Msun)')
plt.scatter(r2[-1], m2[-1], label=f'11 Msun progenitor: ({r2[-1]:.6f} Rsun, {m2[-1]:.6f} Msun)')
plt.grid()
plt.xlim(0)
plt.ylim(0)
plt.legend()
plt.show()
print(f'{r1[-1]:.6e}, {r2[-1]:.6e}')
```
